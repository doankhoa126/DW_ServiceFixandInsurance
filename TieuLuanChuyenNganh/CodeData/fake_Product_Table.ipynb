```python
import csv
import random

list_ins=['1','2','3']

with open('laptop_price.csv',"rt") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    next(csv_reader)
    for row in csv_reader:
        Product_ID= row['Product_ID']
        Supplier_ID= row['Supplier_ID']
        ProductType_ID= row['ProductType_ID']
        Product_Name= row['Product_Name']
        OS = row['OS']
        Screen = row['Screen']
        CPU = row['CPU']
        Ram = row['Ram']
        Memory = row['Memory']
        Price = row['Price']
        

        
        id_Product =f'Prt{Product_ID}'
        Insurance_Time=random.choice(list_ins)
        query = "Insert Into Dim_Product(Product_ID,Supplier_ID,ProductType_ID,Product_Name,OS,Screen,CPU,Ram,Memory,Price,Insurance_Time) values('{Product_ID}','{Supplier_ID}','{ProductType_ID}','{Product_Name}','{OS}','{Screen}','{CPU}','{Ram}','{Memory}',{Price},{Insurance_Time});".format(Product_ID=id_Product,Supplier_ID=Supplier_ID,ProductType_ID=ProductType_ID ,Product_Name=Product_Name, OS = OS, Screen = Screen, CPU=CPU,Ram=Ram,Memory=Memory,Price=Price,Insurance_Time=Insurance_Time)
        print(query)
        
        
```
