```python
import csv
import random

list_type=['PT01']
with open('phonelinhkien.csv',"rt") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    next(csv_reader)
    for row in csv_reader:
        
        Component_ID= row['Component_Names']
        Component_Name= row['Component_Name']
        Component_Price = row['Component_Price']
        
        
        ProductType_ID=random.choice(list_type)
        
        query = "Insert Into Dim_Component(Component_ID,ProductType_ID,Component_Name,Component_Price) values('{Component_ID}','{ProductType_ID}',N'{Component_Name}',{Component_Price});".format(Component_ID=Component_ID, ProductType_ID=ProductType_ID, Component_Name=Component_Name ,Component_Price=Component_Price)
        print(query)
        
        
```
